# Training-Trend-Display (For YOLOv10)



### The training process file from the YOLO training download allows to use this script to visualise the change in training set and validation set losses during the output training process.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path ='/Users/piglet/Desktop/Results/Detection/RAFDB/v10_RAFDB_YOLO_Detection/train/results.csv'
data = pd.read_csv(file_path)

data.columns = data.columns.str.strip()

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)  
plt.plot(data['epoch'], data['train/box_om'], label='Train Box Loss', linewidth=2)
plt.plot(data['epoch'], data['train/cls_om'], label='Train Class Loss', linewidth=2)
plt.plot(data['epoch'], data['train/dfl_om'], label='Train DFL Loss', linewidth=2)
plt.title('Training Loss Over Epochs', fontsize=14, fontweight='bold')
plt.xlabel('Epoch', fontsize=12, fontweight='bold')
plt.ylabel('Loss', fontsize=12, fontweight='bold')
plt.legend(fontsize=12, title_fontsize='13', fancybox=True)
plt.grid(True)

plt.subplot(1, 2, 2) 
plt.plot(data['epoch'], data['val/box_om'], label='Validation Box Loss', linewidth=2)
plt.plot(data['epoch'], data['val/cls_om'], label='Validation Class Loss', linewidth=2)
plt.plot(data['epoch'], data['val/dfl_om'], label='Validation DFL Loss', linewidth=2)
plt.title('Validation Loss Over Epochs', fontsize=14, fontweight='bold')
plt.xlabel('Epoch', fontsize=12, fontweight='bold')
plt.ylabel('Loss', fontsize=12, fontweight='bold')
plt.legend(fontsize=12, title_fontsize='13', fancybox=True)
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### The training process file from the YOLO training download allows to use this script to visualise the change in all mAP and learning rate during metrics the output training process.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/piglet/Desktop/实验结果/Detection/RAFDB/v10_RAFDB_YOLO_Detection/train/results.csv'  

data = pd.read_csv(file_path)
data.columns = data.columns.str.strip()
epochs = data['epoch']
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

for col in data.columns:
    if 'mAP' in col:
        ax1.plot(epochs, data[col], label=col)
ax1.set_title('mAP Changes over Epochs')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('mAP')
ax1.legend()

for col in data.columns:
    if 'lr/' in col:
        ax2.plot(epochs, data[col], label=col)
ax2.set_title('Learning Rate Changes over Epochs')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Learning Rate')
ax2.legend()

fig.tight_layout()
plt.show()
```
